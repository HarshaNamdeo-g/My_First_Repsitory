## Data Cleaning
1. Check null values
2. isna() # this returns boolean output like it will return true for nan(Not A Number)and false for not nan

```python
df.isna()
```

```python
df.isna().sum() # we can use aggrigate functions like this instead to get the sensible output
```

```python
# All rows of column "Precip Type" which contains nan value 
df[df["Precip Type"].isna()==True] # By default it is true otherwise False will return all records without nan values
```

```python
dfclean=df[df["Precip Type"].isna()==False] # WE Transferred the clean data into a new data frame "dfclean"
dfclean.isna().sum()
```

```python
# Find out the duplicate rows
df.duplicated()
```

```python
df.duplicated().sum()
```

```python
dfclean.duplicated().sum() #
```

```python
df[df.duplicated()==True].head()# This will show all the duplicated rows in dataframe df
```

```python
# Remove the duplicated rows from new dataframe dfclean
# dfclean=dfclean.drop_duplicates(keep='first') one way to update the dataframe
dfclean.drop_duplicates(keep='first',inplace=True)# Another way of updating is passing another parameter which is 'inplace = True'
```

```python
dfclean.duplicated().sum() #our updated dataframe has no duplicate rows now
```
