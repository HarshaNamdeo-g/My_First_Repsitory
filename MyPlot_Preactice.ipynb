```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,10)
y=np.arange(1,10)
a=[2,3,4,5]
b=[3,5,6,8]
plt.plot(a,b,color="red",label="data2")   
plt.plot(x,y,label="data1")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```
